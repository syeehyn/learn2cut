```python
import pandas as pd
import numpy as np
from glob import glob
import json
```

```python
!ls
```

```python
lst = [json.load(open(i)) for i in glob('./*easy.json')]
observations = []
for i in lst:
    try:
        observations.append(json.loads(i))
    except:
        observations.append(i)
        
idx = [glob('./*easy.json')]
pd.DataFrame(np.array(observations), index = idx)#.T.plot(figsize=(20, 15))
```
